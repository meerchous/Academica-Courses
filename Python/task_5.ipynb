```python
'''1. Рисование прямоугольника из звездочек:
Задача: Напишите программу, которая рисует прямоугольник из звездочек (*).
Пользователь вводит высоту и ширину прямоугольника (по очереди примите два значения от пользователя, не одновременно).
Инструменты: Используйте внешний цикл for для перебора строк и внутренний цикл for для перебора столбцов в каждой строке.
Используйте print
Вывод:
 - Предположим, пользователь вводит высоту 3 и ширину 5. Тогда вывод программы будет следующим:
     *****
     *****
     *****'''

a, b = int(input('Высота: ')), int(input('Ширина: '))
for _ in range(a):
    for _ in range(b):
        print('*',end='')
    print()
    
# for _ in range(a):
#     print('*'*b)
```

```python
'''1. Создание простого счетчика времени:
Задача: Напишите программу, которая ведет обратный отсчет от заданного пользователем времени в секундах до нуля.
Инструменты: Используйте цикл while для уменьшения счетчика каждую секунду и вывода оставшегося времени.
Программа должна остановиться, когда счетчик достигнет нуля.
Вывод:
- Если пользователь задает время 5 секунд, то вывод будет выглядеть так:
     5
     4
     3
     2
     1
     Время истекло!'''

n = int(input())
while n>0:
    print(n)
    n -= 1
else:
    print('Время истекло!')

```

```python
'''2.Программа угадывания числа:
Задача: Напишите программу, где компьютер загадывает число от 1 до 100, а пользователь пытается его угадать.
После каждой попытки программа должна сообщать, больше или меньше загаданное число, чем введенное пользователем.
Процесс продолжается до тех пор, пока число не будет угадано.
Инструменты: Используйте цикл while для повторения действий, пока пользователь не угадает число.
Вывод:
- Представим, что компьютер загадал число 42, а пользователь вводит следующие числа: 50, 30, 45, 42. Тогда вывод будет таким:
     Ваше число: 50
     Загаданное число меньше!
     Ваше число: 30
     Загаданное число больше!
     Ваше число: 45
     Загаданное число меньше!
     Ваше число: 42
     Вы угадали число!'''

import random as r

x = r.randint(1,100)
i = 0
while x!=i:
    i = int(input('Ваше число: '))
    if i>x:
        print('Загаданное число меньше!')
    elif i<x:
        print('Загаданное число больше!')
    else:
        print('Вы угадали число!')
        break
```
