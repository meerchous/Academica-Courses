**Лекция: Использование функции Pivot в Pandas**

---

### Введение в Pivot

Функция `pivot` в Pandas предназначена для переформатирования или изменения ориентации данных в `DataFrame`. Это мощный инструмент для переструктурирования данных, который позволяет легко изменять формат таблицы, делая данные более читаемыми и удобными для анализа.

---

### Основы использования Pivot

**Синтаксис:**

```python
DataFrame.pivot(index=None, columns=None, values=None)
```

- `index`: имя или список имен для создания новых индексов в результирующем DataFrame.
- `columns`: имя или список имен, которые станут новыми столбцами.
- `values`: имя или список имен столбцов, которые будут заполнять ячейки нового DataFrame.

---

### Пример использования

Допустим, у нас есть следующий DataFrame, который содержит данные о продажах фруктов в разные дни:

```python
import pandas as pd

data = pd.DataFrame({
    'Date': ['2020-01-01', '2020-01-01', '2020-01-02', '2020-01-02'],
    'Fruit': ['Apple', 'Banana', 'Apple', 'Banana'],
    'Quantity': [5, 7, 8, 3]
})
```

**Применение Pivot:**

Мы хотим переориентировать данные так, чтобы даты были индексами, фрукты - столбцами, а значениями было количество проданных фруктов.

```python
pivot_df = data.pivot(index='Date', columns='Fruit', values='Quantity')
print(pivot_df)
```

Результатом будет `DataFrame`, где каждая строка соответствует определенной дате, каждый столбец - определенному фрукту, а значения в таблице показывают количество проданных фруктов.

---

### Расширенные применения

- **Работа с большими наборами данных:** Pivot особенно полезен при работе с большими наборами данных, где необходимо агрегировать и сравнивать данные по разным категориям.
- **Визуализация данных:** После применения pivot данные часто становятся более удобными для визуализации, например, с использованием библиотеки Seaborn или Matplotlib.

---

### Важные замечания

- **Уникальность индекса/столбца:** Комбинация значений в столбцах `index` и `columns` должна быть уникальной. В противном случае операция pivot завершится ошибкой.
- **Обработка отсутствующих данных:** Если в данных есть пропуски, в результирующем DataFrame они будут отображаться как `NaN`.

---

### Заключение

Функция pivot в Pandas является важным инструментом для преобразования и агрегирования данных. Она позволяет эффективно перестраивать и суммировать данные для анализа и визуализации, делая работу с данными более гибкой и интуитивно понятной.

### Проект: Анализ данных о пассажиропотоке в авиации

**Цель проекта:** Исследовать данные о пассажиропотоке авиакомпаний, чтобы выявить сезонные тренды и изменения в пассажиропотоке со временем.

**Шаг 1: Загрузка данных**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка датасета
flights = sns.load_dataset('flights')
print(flights.head())
```

**Шаг 2: Визуализация количества пассажиров по годам**

Для начала мы можем визуализировать общее количество пассажиров по годам.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=flights, x='year', y='passengers', estimator=sum)
plt.title('Общее количество пассажиров по годам')
plt.show()
```

**Шаг 3: Анализ сезонных трендов**

Теперь давайте исследуем, как меняется количество пассажиров в зависимости от месяца.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=flights, x='month', y='passengers', ci=None)
plt.title('Среднее количество пассажиров по месяцам')
plt.show()
```

**Шаг 4: Тепловая карта пассажиропотока**

Тепловая карта поможет нам лучше понять изменения пассажиропотока по месяцам и годам.

```python
# Подготовка данных для тепловой карты
flights_pivot = flights.pivot("month", "year", "passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", linewidths=.5, cmap='coolwarm')
plt.title('Тепловая карта пассажиропотока по месяцам и годам')
plt.show()
```

**Шаг 5: Парные отношения**

Парные отношения позволяют нам исследовать связь между количеством пассажиров и годами для каждого месяца.

```python
sns.pairplot(data=flights, hue='month')
plt.show()
```
