```python
'''1. Расчет Суммы Чисел: Написать программу, которая использует цикл for для вычисления суммы всех чисел от 1 до 100.'''

count = 0
for i in range(1,101):
    count += i 
print(count)
```

```python
'''2. Таблица Умножения: Используя цикл for, создать программу, которая выводит таблицу умножения для выбранного пользователем числа от 1 до 10. Например, если пользователь выберет число 5, программа должна вывести таблицу умножения на 5, то есть вот такой вывод:
5 * 1 = 5
5 * 2 = 10
5 * 3 = 15
5 * 4 = 20
5 * 5 = 25
5 * 6 = 30
5 * 7 = 35
5 * 8 = 40
5 * 9 = 45
5 * 10 = 50'''

num = int(input('Число от 1 до 10:'))
for i in range(1,11):
    print(num,'*',i,'=',num*i )
```

```python
'''3. Генерация Списка Четных Чисел: Написать программу, использующую цикл for, для создания и вывода списка всех четных чисел от 2 до 20.'''

# for i in range(2,21,2):
#     print(i, end=' ')

for i in range(2,21):
    if i%2==0:
        print(i, end=' ')
    else:
        continue
```

```python
'''4. Обратный Отсчет: Создать программу, которая использует цикл for для вывода обратного отсчета от 10 до 0, после чего выводит сообщение "Пуск!".'''

for i in range(10,-1,-1):
    print(i)
    
else: 
    print('Пуск!')
```

```python
'''5. Работа со Строками: Написать программу, которая принимает строку от пользователя и использует цикл for для вывода каждого символа этой строки в столбик.
Например, для строки "Привет" программа должна вывести:
П
р
и
в
е
т
Для решения задачи "Работа со Строками", где необходимо вывести каждый символ введенной строки в столбик, вы можете следовать таким шагам:
- Получение Строки от Пользователя: Сначала нужно запросить у пользователя ввести строку.
Это может быть реализовано с помощью функции input() в Python.
- Использование Цикла for: Далее используйте цикл for для перебора каждого символа в строке.
В Python строка может быть непосредственно использована в цикле for для итерации по её символам.
- Вывод Символов: В теле цикла выведите текущий символ.
Функция print() в Python по умолчанию добавляет символ новой строки после каждого вызова, так что каждый символ будет автоматически выводиться на новой строке.'''

s = input('Напиши что нибудь: ')
for i in s:
    print(i)
```
