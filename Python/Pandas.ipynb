**Лекция: Введение в Pandas для анализа данных**

---

### Введение

Pandas — это высокоуровневая библиотека анализа данных в Python, которая обеспечивает широкие возможности для работы с табличными данными. Она предлагает удобные структуры данных для эффективного манипулирования и анализа данных.

---

### Основные структуры данных

1. **DataFrame**:
   - Двумерная таблица с метками на осях.
   - Может содержать разнотипные данные.
   - Подходит для представления реальных данных, например, таблиц из Excel, SQL-запросов и т.д.

2. **Series**:
   - Одномерный массив с метками оси.
   - Может содержать любой тип данных.

---


`Series` и `DataFrame` являются основными структурами данных в библиотеке Pandas для Python, и они тесно связаны между собой, но используются для разных целей в анализе данных.

### Series

1. **Одномерность:**
   - `Series` представляет собой одномерный массив с индексированными данными. Можно представить его как один столбец данных.

2. **Структура данных:**
   - Каждый элемент в `Series` имеет уникальный индекс, который используется для доступа к данным. Индексы могут быть числовыми, строковыми или временными метками.

3. **Гомогенность данных:**
   - Все данные в `Series` должны быть одного типа (например, целые числа, вещественные числа, строки и т.д.).

4. **Создание:**
   - `Series` создается из списка, массива или скаляра с помощью функции `pd.Series()`.

   Пример:
   ```python
   import pandas as pd

   series = pd.Series([1, 2, 3, 4, 5])
   ```

### DataFrame

1. **Двумерность:**
   - `DataFrame` представляет собой двумерную таблицу с индексированными строками и столбцами. Это похоже на таблицу в Excel или SQL-таблицу.

2. **Структура данных:**
   - `DataFrame` состоит из одного или нескольких `Series`, которые разделяют один и тот же индекс. Каждый столбец в `DataFrame` является `Series`.

3. **Гетерогенность данных:**
   - В отличие от `Series`, `DataFrame` может содержать данные различных типов (например, целые числа, строки, вещественные числа и т.д.) в разных столбцах.

4. **Создание:**
   - `DataFrame` создается из словаря списков, массивов, других `DataFrame` или из `Series` с помощью функции `pd.DataFrame()`.

   Пример:
   ```python
   df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
   ```

### Выводы

- Используйте `Series`, когда работаете с одномерным массивом.
- Используйте `DataFrame`, когда вам нужна табличная структура с несколькими столбцами разных типов данных.
- `DataFrame` является более сложной и мощной структурой по сравнению с `Series` и чаще используется в анализе данных.

В Pandas для одновременного выбора строк и столбцов из `DataFrame` обычно используются `.loc[]` и `.iloc[]`. Эти индексаторы позволяют вам эффективно выбирать данные по меткам или по их числовым позициям.

### Использование `.loc[]` для выбора по меткам

`.loc[]` используется для выбора строк и столбцов по их меткам. Синтаксис:

```python
df.loc[rows, columns]
```

- `rows` - метки строк, которые вы хотите выбрать.
- `columns` - метки столбцов, которые вы хотите выбрать.

**Пример:**

```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}, index=['one', 'two', 'three'])

# Выбор строк 'one' и 'two', столбцов 'A' и 'B'
result = df.loc[['one', 'two'], ['A', 'B']]
print(result)
```

### Использование `.iloc[]` для выбора по числовым позициям

`.iloc[]` используется для выбора строк и столбцов по их числовым индексам. Синтаксис:

```python
df.iloc[rows, columns]
```

- `rows` - числовые индексы строк.
- `columns` - числовые индексы столбцов.

**Пример:**

```python
# Выбор первых двух строк и первых двух столбцов
result = df.iloc[0:2, 0:2]
print(result)
```

Оба эти метода (`loc` и `iloc`) очень мощные и предоставляют гибкость при выборе данных из `DataFrame`. Важно помнить, что `loc` использует метки для выбора, в то время как `iloc` использует позиции на основе целых чисел.

### Использование `dropna` с параметром `subset`

Когда вы используете параметр `subset` в `dropna`, вы указываете столбцы, для которых нужно проверять наличие `NaN`. Если `NaN` обнаружен в одном из этих столбцов, соответствующая строка удаляется.

**Пример:**

Предположим, у вас есть следующий `DataFrame`:

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, 2, 3, np.nan]
})
```

Если вы хотите удалить строки, в которых есть `NaN` только в столбце 'A', вы можете сделать следующее:

```python
df_cleaned = df.dropna(subset=['A'])
print(df_cleaned)
```

Этот код удалит только те строки, где в столбце 'A' есть пропущенные значения. Строки, в которых пропущенные значения находятся в других столбцах ('B' или 'C'), останутся нетронутыми.

### Использование `how`

Параметр `how` определяет, будут ли удаляться строки, если обнаружены `NaN`:

- `how='any'` (по умолчанию): удаляет строки, в которых есть хотя бы одно `NaN` в указанных столбцах.
- `how='all'`: удаляет строки, где все указанные столбцы содержат `NaN`.

**Пример:**

Если вы хотите удалить строки, где вся подмножества столбцов содержат `NaN`:

```python
df_cleaned = df.dropna(subset=['A', 'B'], how='all')
print(df_cleaned)
```

Этот код удалит строки, где столбцы 'A' и 'B' одновременно содержат `NaN`.

Функция `fillna()` в Pandas используется для заполнения пропущенных значений (`NaN`) в DataFrame или Series. Она предлагает гибкость в выборе значения для замены и позволяет заполнять пропущенные данные различными способами, включая использование констант, средних значений, медиан и т.д.

### Заполнение пропущенных значений константой

Вы можете заполнить все пропущенные значения в DataFrame или в определенном столбце фиксированным значением.

**Пример:**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, np.nan, 3],
    'B': [4, 5, np.nan]
})

# Заполнение всех NaN нулями
df_filled = df.fillna(0)
```

### Заполнение пропущенных значений различными значениями для разных столбцов

Вы можете указать разные значения для разных столбцов, используя словарь, где ключ — это имя столбца, а значение — это значение, которым следует заменить `NaN`.

**Пример:**

```python
# Заполнение NaN в столбце 'A' нулями, а в столбце 'B' единицами
df_filled = df.fillna({'A': 0, 'B': 1})
```

### Заполнение пропущенных значений средним значением столбца

Чтобы заполнить пропущенные значения средним (или медианой, модой) по столбцу, сначала вычислите среднее значение столбца, а затем используйте `fillna()`.

**Пример:**

```python
# Заполнение NaN в столбце 'A' средним значением по этому столбцу
df['A'].fillna(df['A'].mean(), inplace=True)

# Или для всего DataFrame
df_filled = df.apply(lambda x: x.fillna(x.mean()))
```

В этом примере `df['A'].mean()` вычисляет среднее значение столбца 'A', а `fillna()` затем использует это значение для замены всех `NaN` в столбце 'A'. Использование `lambda` функции в `apply()` позволяет применить этот подход ко всем столбцам в DataFrame.

### Примечания

- Использование `inplace=True` в методе `fillna()` применит изменения непосредствен

**Лекция: Использование GroupBy в Pandas**

---

### Введение в GroupBy

Функция `groupby` в Pandas является одной из наиболее мощных и гибких возможностей для агрегирования, трансформации и анализа данных. Она позволяет группировать данные на основе некоторых критериев и затем применять агрегирующие функции к каждой группе независимо.

---

### Основной принцип работы

Процесс работы с `groupby` обычно включает три шага: разделение данных на группы, применение функции к каждой группе и объединение результатов.

1. **Разделение:** Данные делятся на группы на основе одного или нескольких ключей.
2. **Применение:** К каждой группе применяется функция, которая может быть агрегированием, трансформацией или фильтрацией.
3. **Объединение:** Результаты этих операций объединяются в новую структуру данных.

---

### Пример использования GroupBy

Допустим, у нас есть DataFrame с данными о продажах в разных магазинах.

```python
import pandas as pd

data = pd.DataFrame({
    'Shop': ['Shop1', 'Shop1', 'Shop2', 'Shop2', 'Shop3'],
    'Product': ['Apples', 'Oranges', 'Apples', 'Oranges', 'Apples'],
    'Sales': [10, 15, 5, 7, 6]
})
```

#### Группировка данных

```python
grouped = data.groupby('Shop')
```

#### Агрегирование данных

```python
# Суммирование продаж по магазинам
total_sales = grouped['Sales'].sum()
print(total_sales)
```

#### Комбинирование нескольких агрегаций

```python
# Использование agg для комбинирования агрегаций
sales_stats = grouped['Sales'].agg([sum, min, max])
print(sales_stats)
```

**Лекция: Использование метода value_counts в Pandas**

---

### Введение в value_counts

Метод `value_counts` в Pandas - это удобный инструмент для подсчета количества уникальных значений в столбце `Series`. Этот метод чрезвычайно полезен для анализа данных, особенно при работе с категориальными данными, так как он позволяет быстро оценить распределение данных.

---

### Основные особенности

- **Простота использования:** Метод `value_counts` применяется к объекту `Series` и возвращает `Series`, где индексы — это уникальные значения, а значения — это соответствующие количества их появлений в исходном `Series`.
- **Сортировка по умолчанию:** Результаты по умолчанию отсортированы по убыванию, т.е. самые часто встречающиеся значения будут на вершине.
- **Настройка сортировки:** Можно изменить порядок сортировки или отсортировать результаты по индексу.
- **Нормализация:** Поддерживает возможность нормализации результатов для получения долей вместо количеств.

---

### Примеры использования

Предположим, у нас есть DataFrame с данными о пассажирах на Титанике:

```python
import pandas as pd

# Загрузка данных о пассажирах Титаника
data = pd.read_csv('titanic.csv')

# Подсчет количества пассажиров в каждом классе
class_counts = data['Pclass'].value_counts()
print(class_counts)
```

#### Сортировка по индексу

```python
class_counts_sorted = data['Pclass'].value_counts().sort_index()
print(class_counts_sorted)
```

#### Нормализация результатов

```python
class_distribution = data['Pclass'].value_counts(normalize=True)
print(class_distribution)
```

Этот код вернет долю пассажиров в каждом классе от общего числа пассажиров.

#### Выборка с использованием value_counts

```python
# Топ 5 самых частых имен пассажиров
top_names = data['Name'].value_counts().head(5)
print(top_names)
```
