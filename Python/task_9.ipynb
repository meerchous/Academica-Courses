с. Создайте одномерный массив из 8 случайных целых чисел в диапазоне от 1 до 20.



```python
import numpy as np

a = np.arange(5); print(a,'\n')
b = np.random.randint(10,size=(3,3,3)); print(b,'\n')
c = np.random.randint(10,size=8)

for i in range(len(c)):
    c[i] = np.random.randint(1,21)
    
print(c)
```

```python
import cv2
import matplotlib.pyplot as plt
from skimage import data, color

image = cv2.imread('WWS.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

blue_white_image = np.copy(image)
blue_white_image[:,:,0] = 10 # Red
blue_white_image[:,:,1] = 10 # Green

bright_areas = blue_white_image[:,:,2] > 120
blue_white_image[bright_areas,2] = 255

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blue_white_image)
plt.title('Blue and White Image')
plt.axis('off')

plt.show()
```
