Задание 1

Постройте линейную диаграмму, отображающую температуру в городе за одну неделю. Создайте два списка чисел, один для оси X, представляющий дни недели, а другой для оси Y, представляющий температуру. Добавьте подписи к осям и заголовок графика.

```python
import matplotlib.pyplot as plt 

weekdays = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
temp = [-8,-3,3,2,1,-5,3]


plt.plot(weekdays,temp, 'ro-')
plt.title('Weather Forecast')
plt.xlabel('Weekdays')
plt.ylabel('Temperature °C')
plt.grid()
plt.show()
```

Постройте точечную диаграмму, иллюстрирующую взаимосвязь между количеством часов учёбы и оценками студентов. Создайте два списка чисел, один для оси X, представляющий часы учебы, а другой для оси Y, представляющий оценки студентов. Добавьте подписи к осям и заголовок.



```python
study_hours = [8,7,8,6,5,4,5,6]
marks = [98,85,90,79,67,55,64,83]
plt.scatter(study_hours,marks)
plt.title('Students study hours&marks correlation')
plt.xlabel('Study hours')
plt.ylabel('Marks')
plt.grid()
plt.show()
```

Задание 4

Создайте гистограмму для анализа частоты встречаемости различных оценок. Создайте набор данных, используя генерацию случайных чисел. Например, оценки могут быть в диапазоне от 2 до 5 (включительно). Для этого используйте библиотеку numpy (функцию numpy.random.randint) Выберите подходящее количество бинов для гистограммы. Например, для оценок может подойти 8 бина. Добавьте подписи к осям и заголовок.



```python
import numpy as np

marks = np.random.randint(low = 2, high = 6, size = 10)

plt.hist(marks,bins=8)
plt.title('Frequency of occurrence of various estimates')
plt.xlabel('Оценки')
plt.ylabel('Частота встречаемости')
plt.show()
```

Задание 5

Создайте две Столбчатые Диаграммы на одном изображении:

Первая Диаграмма: Постройте столбчатую диаграмму, показывающую количество чашек разных видов кофе (например, эспрессо, латте, капучино), проданных в кофейне за день.
Вторая Диаграмма: Создайте столбчатую диаграмму для пяти различных закусок (например, круассаны, багеты, печенья и т.д.), также проданных в этот же день.
Добавьте подписи к осям и заголовок.

```python
plt.subplot(1,2,1)
plt.bar(['Latte','Espresso','Capuccino','Americano','Mocha'],np.random.randint(1,50,5))
plt.title('Виды кофе проданных за день')
plt.xlabel('Coffee types')
plt.ylabel('Sales')


plt.subplot(1,2,2)
plt.bar(['croissants','baguettes', 'cookies', 'chocolate', 'cheesecake'],np.random.randint(1,50,5))
plt.title('Виды закусок проданных за день')
plt.xlabel('Snack types')
plt.ylabel('Sales')

plt.show()
```

```python
# Альтернативный вариант

fig, ax = plt.subplots(1,2, figsize=(20,5))

ax[0].bar(['Latte','Espresso','Capuccino','Americano','Mocha'],np.random.randint(1,50,5))
ax[0].title.set_text('Виды кофе проданных за день')
ax[0].set_xlabel('Coffee types')
ax[0].set_ylabel('Sales')

ax[1].bar(['croissants','baguettes', 'cookies', 'chocolate', 'cheesecake'],np.random.randint(1,50,5))
ax[1].title.set_text('Виды закусок проданных за день')
ax[1].set_xlabel('Snack types')
ax[1].set_ylabel('Sales')

plt.show()
```
