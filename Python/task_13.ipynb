2. Удалите колонку 'LikePresentations', она нам не понадобится

```python
df.drop('LikePresentation',axis=1,inplace=True)
df
```

3. Измените названия ваших колонок - разделите слова в названии колонок. Например, замените AcademicPerformance на Academic Performance

```python
df.rename(columns={'Age':'age','Gender':'Sex','AcademicPerformance':'Academic Performance','TakingNoteInClass':'Taking Note In Class','DepressionStatus':'Depression Status', 'FaceChallangesToCompleteAcademicTask':'Face Challanges To Complete Academic Task','SleepPerDayHours':'Sleep Per Day Hours','NumberOfFriend':'Number Of Friend','LikeNewThings':'Like New Things'},inplace=True)
df
```

4.Очистка данных. Удалите дубликаты, если они есть и строки с пропущенными значениями

```python
df.drop_duplicates(inplace=True)
df
```

5.Выявить среднее количество сна студентов

```python
print(f'Среднее количество сна студентов: {df["Sleep Per Day Hours"].mean()}')
```
