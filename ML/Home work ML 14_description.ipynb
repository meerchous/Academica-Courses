2. Прочитайте данные

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Mall_Customers.csv')
print(df.to_string())
```

3. Проверить, есть ли нулевые значения?

```python
df.isna().sum()
```

4. Используйте столбцы "Annual Income (k$)" и "Spending Score (1-100)" в качестве вектора признаков.

```python
X = df[['Annual Income (k$)','Spending Score (1-100)']].values
```

5. Запустите алгоритм Kmeans, где k будет от 1 до 50, и постройте график метода Elbow.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

cs = []
for i in range(1,50):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 49, random_state = 21)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 50), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

6. Каково наилучшее значение K? Найдя лучшее значение для K, сделайте прогноз

```python
kmeans = KMeans(n_clusters = 5, random_state = 21)
kmeans.fit(X)
prediction = kmeans.fit_predict(X)
prediction
```

7. Визуализируйте эти кластеры в 2D-пространстве.

```python
import matplotlib.pyplot as plt

df['target'] = prediction

df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
df3 = df[df.target == 3]
df4 = df[df.target == 4]

plt.figure(figsize = (9,6))
plt.scatter(df0['Annual Income (k$)'],df0['Spending Score (1-100)'], label = '0')
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'], label = '1')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'], label = '2')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'], label = '3')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'], label = '4')
plt.scatter(kmeans.cluster_centers_[: , 0] ,  kmeans.cluster_centers_[: , 1] , color = "magenta" , marker = "*"  , label =  "Centroids")
plt.legend(loc='upper right'  , bbox_to_anchor=(1.20, 1.015))
plt.title("Clusters")
plt.xlabel("Spending Score")
plt.ylabel("Annual Income")
plt.show()
```
