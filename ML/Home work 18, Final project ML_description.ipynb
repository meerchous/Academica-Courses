5. Удалите столбец «id»

```python
data = data.drop(columns='id')
data
```

6. Найдите дубликаты и удалите их

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data
```

7. Определите значение «диагноз» и преобразуйте их в числовую форму.

```python
data['diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data['diagnosis'] = encoder.fit_transform(data['diagnosis'])
data
```

```python
data['diagnosis'].value_counts()
```

8. Используйте столбец «диагноз» как значение y, а остальные столбцы как значение X. Затем нормализуйте вектор признаков.

```python
from sklearn.preprocessing import StandardScaler

X = data.iloc[:,1:]
y = data.iloc[:,0]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

9. Разделите данные на обучающий и тестовый наборы, где 80% будут использоваться для обучения, а остальные — для тестирования. Затем обучите модель

```python
from sklearn.model_selection import train_test_split

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=21)
```

10. Сделайте прогноз

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=21)
model.fit(X_train_scaled, y_train)
prediction = model.predict(X_test_scaled)
```

11. Рассчитайте точность

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(f"Report:\n{classification_report(y_test, prediction)}")
print(f"Confusion matrix:\n{confusion_matrix(y_test, prediction)}")
print(f"Accuracy:\n{accuracy_score(y_test, prediction)}")
```
