**Задача 1: Линейная регрессия**

**Создание искусственного набора данных**: Используйте numpy для генерации случайных данных: независимая переменная x должна быть равномерно распределена в диапазоне от 0 до 20, а зависимая переменная y должна быть сгенерирована по линейной функции y=4x+20 с добавлением нормально распределенной случайной ошибки с математическим ожиданием 0 и стандартным отклонением 5.

```python
import numpy as np

np.random.seed(21)
X = np.arange(0,21)
y = 4 * X + 20 + np.random.normal(0,5,21)
y
```

**Визуализация данных**: Отобразите сгенерированные данные на графике с помощью matplotlib.pyplot, используя функцию scatter для точек данных и plot для линии регрессии.



```python
import matplotlib.pyplot as plt

plt.scatter(X,y,label = 'значения')
plt.plot([0,21],[18,104],color='red',linestyle='--',label='лучшая линия')
plt.title('Линии регрессии')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

**Обучение модели линейной регрессии**: Примените модель линейной регрессии из sklearn.linear_model для обучения на сгенерированных данных. Используйте метод fit для обучения модели, предварительно преобразовав x в двумерный массив.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X.reshape(21,1),y)

predictions = model.predict(X.reshape(21,1))
predictions
```

**Оценка качества модели**: Рассчитайте среднеквадратичную ошибку (MSE) между истинными значениями y и значениями, предсказанными моделью, используя функцию mean_squared_error из sklearn.metrics.



Выведите коэффициенты обученной модели (угловой коэффициент и свободный член). Проанализируйте полученные результаты.

```python
from sklearn.metrics import mean_squared_error

MSE_1 = mean_squared_error(y,predictions) ** (1/2)
coef_1 = model.coef_
intercept_1 = model.intercept_
print(f"Среднеквадратичная ошибка(MSE) 1:{MSE_1}")
print(f"Угловой коэффициент 1:{coef_1}")
print(f"Свободный член 1:{intercept_1}")
```

Убедитесь, что x генерируется как равномерно распределенные значения в диапазоне от 0 до 20.



**Визуализация нового набора данных**: Отобразите сгенерированные точки данных на графике с помощью функции scatter из matplotlib.pyplot.



**Обучение новой модели линейной регрессии**: Используйте класс LinearRegression из sklearn.linear_model для обучения модели на новом наборе данных. Преобразуйте x в двумерный массив перед обучением модели.



**Оценка качества новой модели**: Рассчитайте среднеквадратичную ошибку (MSE) между истинными значениями y и значениями, предсказанными новой моделью. Используйте функцию mean_squared_error из sklearn.metrics.



**Сравнение качества двух моделей линейной регрессии**: Сравните MSE новой модели с MSE модели из первого задания. Проанализируйте, как изменение коэффициентов линейного уравнения (наклона и свободного члена) влияет на качество модели.

```python
np.random.seed(21)
X = np.random.uniform(0,20,100)
y = -2 * X + 5 + np.random.normal(0,5,100)

plt.scatter(X,y,label='значения')
plt.plot([0,20],[8,-39], color='red', linestyle='--',label='лучшая линия')
plt.title('Сгенерированные точки данных')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X.reshape(100,1),y)

predictions = model.predict(X.reshape(100,1))

MSE_2 = mean_squared_error(y,predictions) ** (1/2)
coef_2 = model.coef_
intercept_2 = model.intercept_
print(f"Среднеквадратичная ошибка(MSE) 2:{MSE_2}")
print(f"Угловой коэффициент 2:{coef_2}")
print(f"Свободный член 2:{intercept_2}")
```

**Ответ**
У первой модели среднеквадратичная ошибка меньше. Соответственно она будет лучше прогнозировать. Коэффициенты влияют на линию наклона и на пересечение данных с линий регресии.

**Задача 3: Линейная регрессия с множеством признаков**

Используйте pd.DataFrame(np.random.rand(20, 5)) для генерации DataFrame с 20 наблюдениями и 5 признаками.

**Создайте целевую переменную y по формуле**: y = 4*X[0] - 1.5*X[1] + 2.5*X[2] - 3*X[3] + 0.5*X[4] + np.random.normal(0, 2, 20). Это изменение позволит исследовать влияние новой комбинации коэффициентов на результат.

**Обучение модели линейной регрессии**: Обучите модель линейной регрессии на вашем DataFrame для предсказания y.

**Оценка качества модели**: Рассчитайте среднеквадратичную ошибку (MSE) между реальными и предсказанными значениями y.

```python
import pandas as pd
np.random.seed(21)
X = pd.DataFrame(np.random.rand(20,5))
y = 4*X[0] - 1.5*X[1] + 2.5*X[2] - 3*X[3] + 0.5*X[4] + np.random.normal(0,2,20)

model = LinearRegression()
model.fit(X,y)

predictions = model.predict(X)
MSE_predict = mean_squared_error(y,predictions) ** (1/2)
print(f"Среднеквадратичная ошибка(MSE_predict):{MSE_predict}")

tendency_predict = 4*X[0] - 1.5*X[1] + 2.5*X[2] - 3*X[3] + 0.5*X[4]
MSE_real = mean_squared_error(y, tendency_predict) ** (1/2)
print(f"Среднеквадратичная ошибка(MSE_real):{MSE_real}")
```
