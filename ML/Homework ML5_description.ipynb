```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
train.dropna(inplace=True)
X_train = np.array(train['x'])
y_train = np.array(train['y'])
print(X_train)
print(y_train)
```

```python
test = pd.read_csv('test.csv')
test.dropna(inplace=True)
X_test = np.array(test['x'])
y_test = np.array(test['y'])
print(X_test)
print(y_test)
```

2. Используя данные обучения, нарисуйте график scatter plot.

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, label='data')
plt.title('Linear regression by hand')
plt.legend()
plt.show()
```

3. Обучите модель, используя модели scikit (LinearRegression)

```python
from sklearn.linear_model import LinearRegression

X_train = np.reshape(X_train, (699,1))
y_train = np.reshape(y_train, (699,1))
model = LinearRegression()
model.fit(X_train, y_train)
```

4. Протестируйте модель, используя функцию прогнозирования.

```python
X_test = np.reshape(X_test, (300,1))
y_test = np.reshape(y_test, (300,1))
prediction = model.predict(X_test)
```

5. Верните значения a и b в уравнение: y = ax + b

```python
print(f"a:{float(model.coef_)}")
print(f"b:{float(model.intercept_)}")
print(f"y ={float(model.coef_)}x {float(model.intercept_)}")
```

6. Решите эту проблему вручную, где сначала рассчитайте среднее значение X и среднее значение Y для обучающей выборки.

```python
x_avg = np.average(X_train) #xm(avg)
y_avg = np.average(y_train) #ym(avg)
```

7. Вычтите среднее значение X и Y из X и Y: $x - x_{avg}, y - y_{avg}$

```python
x_new = X_train - x_avg  #xi-xm
y_new = y_train - y_avg  #yi-ym
```

8. Найдите следующие значения: $(x - x_{avg})^{2}, (x - x_{avg})(y - y_{avg})$

```python
a_value = x_new**2
b_value = np.multiply(x_new, y_new)
```

9. Используя значения, найденные на шаге 9, вычислите значение b в следующем уравнении: y = ax + b

```python
a = np.sum(b_value) / np.sum(a_value)
a
```

10. Теперь, используя значение b, найдем значение a в уравнении: y = ax + b

```python
b = y_avg - a * x_avg
b
```
