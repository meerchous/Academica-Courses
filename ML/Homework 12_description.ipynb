1. Для вас предоставлены следующие библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn.model_selection import train_test_split , KFold
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

from collections import Counter
```

2. Также вы можете получить набор данных радужной оболочки, используя следующий код

```python
# import iris dataset
iris = datasets.load_iris()
# np.c_ is the numpy concatenate function
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                      columns= iris['feature_names'] + ['target'])
iris_df.head()
```

3. Используйте столбцы: "sepal length, sepal width, petal length, petal width" в качестве вектора признаков и используйте столбец «target» в качестве метки.

```python
X = iris_df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].values
y = iris_df[['target']].values
```

4. Используйте 20% данных для тестирования и остальные для обучения

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

5. Распечатайте количество образцов в train и test наборе.

```python
print(X_train, y_train)
print(X_test, y_test)
```

6. Вы можете использовать нормализованные данные, используя следующий код

```python
scaler= Normalizer().fit(X_train) # the scaler is fitted to the training set
normalized_x_train= scaler.transform(X_train) # the scaler is applied to the training set
normalized_x_test= scaler.transform(X_test) # the scaler is applied to the test set

print(normalized_x_train.shape, normalized_x_test.shape)
print(y_train.shape, y_test.shape)
```

7. Обучите модель с помощью KNN

```python
model = KNeighborsClassifier(n_neighbors=6)
model.fit(normalized_x_train, y_train)
```

8. Сделайте классификацию, используя обученную модель

```python
prediction = model.predict(normalized_x_test)
```

9. Рассчитайте точность

```python
print(f"Accuracy:{accuracy_score(y_test, prediction)}")
```
