#### меры центральной тенденции

```python
sample_data = [1,2,3,3,3,   5,   7,8,8,9,10]
```

**Mean** - это среднее арифметическое всех значений в наборе данных<br>

$\bar{X} = \frac{\sum_{i=1}^{n} X_i}{n}$<br>
где:
- $\bar{X}$ - среднее значение
- $n$ - количество значений в выборке
- $X_i$ - каждое значение в выборке

```python
# среднее значение (mean)
np.mean(sample_data)
```

```python
# арифметический высчет срденего значения
sum(sample_data) / 11
```

**Медиана** - это значение, которое делит упорядоченный набор данных на две равные части

```python
# медиана
np.median(sample_data)
```

**Мода** - это значение (или значения), которое встречается наиболее часто в наборе данных.

```python
# мода
statistics.mode(sample_data)
```

#### меры разброса данных

Мы приехали в новый город и снимаем жилье. Арендодател предложил цену в 145,000 за 1комнатную квартиру.<br>
Зная цены на аренду за этот год в этом городе, мы оценим насколько предложенная цена приемлема.



```python
# сгенерируем случайные цены аренды
np.random.seed(105)
random_numbers = [round(x, -3) for x in np.random.normal(150000, 15000, 1000)]
low_numbers = [round(x, -3) for x in np.random.uniform(60000, 90000, 50)]
high_numbers = [round(x, -3) for x in np.random.uniform(200000, 300000, 50)]

rent_prices = random_numbers+low_numbers+high_numbers
```

```python
print(f'средняя цена аренды: {np.mean(rent_prices)}')
print(f'медианная цена аренды: {np.median(rent_prices)}')
```

Привычные нам средние значения указывают на то что предложенная цена в 145тыс меньше чем средняя цена в городе, 151тыс. Но достаточно ли нам лишь срдених значений, насколько остальные цены близки к среднему значению?

**Стандартное отклонение** - это мера разброса измеряет, насколько каждое значение в наборе данных отклоняется от среднего значения набора.

$\sigma = \sqrt{\frac{\sum_{i=1}^{N}(X_i - \mu)^2}{N}}$

где:
- $\sigma$ - стандартное отклонение
- $N$ - количество значений
- $X_i$ - каждое значение
- $\mu$ - среднее значение

При нормальном распределении 95% всех значений помещаются в 4 стандартных отклонения, 2 слева и 2 справа от среднего значения.

```python
print(f'старндартное отклонение цен составило: {np.std(rent_prices)}')
```

**Размах (range)** -  разница между самым большим и самым маленьким значениями в наборе данных

```python
min_price = min(rent_prices)
max_price = max(rent_prices)

print(f'минимальная цена аренды: {min_price}')
print(f'миаксимальная цена: {max_price}')
print(f'размах цен: {max_price - min_price}')
```

**Квартили** разбивают упорядоченный набор данных на четыре равные по количеству части

Q1 - значение, которое делит нижнюю четверть данных.<br>
Q2 - медиана, значение, которое делит данные пополам<br>
Q3 - значение, которое делит верхнюю четверть данных.

```python
print(f'Первый квартиль: {np.percentile(rent_prices, 25)}')
print(f'Медиана: {np.percentile(rent_prices, 50)}')
print(f'Третий квартиль: {np.percentile(rent_prices, 75)}')
```

**Квартили наглядно**

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(105)  # for reproducibility
random_ages = np.random.randint(10, 81, 20)
random_ages.sort()

q1, q2, q3 = np.percentile(random_ages, [25, 50, 75])

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].scatter(np.arange(1, 21), [1] * 20, c='green', edgecolors='black')
axs[0].set_yticks([])
axs[0].axvline(x=5.5, color='blue', linestyle='--')
axs[0].axvline(x=10.5, color='blue', linestyle='--')
axs[0].axvline(x=15.5, color='blue', linestyle='--')
axs[0].set_xticks(np.arange(1, 21, 1))
axs[0].set_title('распределим людей на 4 равные группы')

axs[1].scatter(np.arange(1, 21), random_ages, c='green', edgecolors='black')
axs[1].set_yticks(np.arange(10, 81, 10))
axs[1].axvline(x=5.5, color='blue', linestyle='--')
axs[1].axvline(x=10.5, color='blue', linestyle='--')
axs[1].axvline(x=15.5, color='blue', linestyle='--')
axs[1].set_xticks([5.5, 10.5, 15.5])
axs[1].set_ylabel('возраст')
axs[1].set_title('учтем их возраст на у-аксисе')

axs[2].scatter(np.arange(1, 21), random_ages, c='green', edgecolors='black')
# axs[2].axvline(x=5.5, color='blue', linestyle='--')
# axs[2].axvline(x=10.5, color='blue', linestyle='--')
# axs[2].axvline(x=15.5, color='blue', linestyle='--')
axs[2].axhline(y=q1, color='red', linestyle='--', label='квартиль')
axs[2].axhline(y=q2, color='red', linestyle='--')
axs[2].axhline(y=q3, color='red', linestyle='--')
axs[2].legend()
axs[2].set_yticks([q1, q2, q3])
axs[2].set_title('найдем квартили возраста людей')
             

plt.tight_layout()

plt.show()
```

**Межквартильный размах (Interquartile Range, IQR)** - представляет собой разницу между верхним (Q3) и нижним (Q1) квартилями набора данных.<br>
IQR предоставляет информацию о разбросе среди средних 50% значений в данных.

```python
IQR = np.percentile(rent_prices, 75) - np.percentile(rent_prices, 25)
print(f'Межквартильный размах цен составил: {IQR}')
```

Квартили и IQR помогли нам понять что половина цен на аренду сконцентрировалась в промежутку от 139тыс до 162 тыс, и такой интервал составил всего 23тыс.

```python
plt.figure(figsize=(12, 8))
plt.hist(rent_prices, bins=np.arange(60000, 310000, 10000), rwidth=0.9)
plt.axvline(np.percentile(rent_prices, 25), color='red', linestyle='--', label='квартили')
plt.axvline(np.median(rent_prices), color='red', linestyle='--')
plt.axvline(np.percentile(rent_prices, 75), color='red', linestyle='--')
# plt.axvline(np.mean(rent_prices), color='black', linestyle='-', label='mean')
plt.legend()
plt.xticks(np.arange(60000, 320000, 20000))
plt.title('Распределение цен на аренду')
plt.xlabel('цены')
plt.ylabel('кол-во квартир')
plt.show()
```

Визуализация данных: **Box Plot (ящик с усами)**<br>
визуальный способ представления распределения данных через показатели центральной тенденции и разброса

```python
# Рассчитываем квартили
median = np.median(rent_prices)
q1 = np.percentile(rent_prices, 25)
q3 = np.percentile(rent_prices, 75)

# Рассчитываем IQR
iqr = q3 - q1

# Визуализация boxplot в вертикальном положении
plt.figure(figsize=(5, 6))
plt.boxplot(rent_prices, vert=True, widths=0.7, patch_artist=False, medianprops={'color':'red'})

# Добавляем текстовые метки для квартилей и IQR
plt.text(1.3, q1, 'Q1', fontsize=10, color='black', fontweight='bold')
plt.text(1.3, median, 'Median', fontsize=10, color='red', fontweight='bold')
plt.text(1.3, q3, 'Q3', fontsize=10, color='black', fontweight='bold')
plt.text(1.1, q1 - 1.5*iqr, 'Q1 - 1.5*IQR', fontsize=10, color='blue', fontweight='bold')
plt.text(1.1, q3 + 1.5*iqr, 'Q3 + 1.5*IQR', fontsize=10, color='blue', fontweight='bold')

# Добавляем заголовок и метки осей
plt.title('Boxplot цен на аренду', fontsize=14)
plt.xticks([1], ['Данные'])
plt.yticks(np.arange(50000, 320000, 20000))
plt.ylabel('Значения', fontsize=12)

# Отображаем график
plt.show()
```

Ящик
- Верхний и нижний края ящика представляют собой третий (Q3) и первый (Q1) квартили соответственно.
- Линия внутри ящика обозначает медиану (Q2) - значение, делящее данные пополам.<br>

Усы
- Усы идут от краев ящика к крайним значениям, которые не являются выбросами.
- Верхний предел усов Q3+1.5×IQR
- Нижний предел усов Q1-1.5×IQR

Выбросы (Outliers)
- Значения за пределами "усов" обычно считаются выбросами и обозначаются отдельными точками или маркерами.

**еще немного о стандартном отклонении**

```python
# пример1: равномерное распределение
np.random.seed(105)
data = np.random.uniform(0, 100, 10000).tolist()

ticks = np.arange(0, 110, 10).tolist()
std = round(np.std(data), 1)

plt.hist(data, bins=ticks, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label=f'Mean')
plt.axvline(np.mean(data) + np.std(data), color='orange', linestyle='dashed', linewidth=2, label=f'Mean + SD ({std})')
plt.axvline(np.mean(data) - np.std(data), color='orange', linestyle='dashed', linewidth=2, label=f'Mean - SD (-{std})')

plt.title('Гистограмма с линией стандартного отклонения', fontsize=14)
plt.xlabel('Значения', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.legend(loc='lower left')

plt.show()
```
