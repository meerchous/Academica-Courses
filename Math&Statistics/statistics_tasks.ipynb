Задача 1 - Вероятность распределения и уровень значимости

В кафе утверждают, что каждая порция эспрессо содержит ровно 30 мл кофе. Однако, на основании многолетних наблюдений, известно, что объем порции подчиняется нормальному распределению со средним значением 30 мл и стандартным отклонением 1 мл. Контролер качества хочет проверить, какова вероятность того, что случайно выбранная порция эспрессо будет содержать менее 29 мл кофе, что считается неприемлемым для стандартов кафе. Используя пороговое значение для уровня значимости 0.05 определите соответствует ли реальная практика утверждениям кафе

```python
from scipy.stats import norm

mean = 30
std = 1
threshold = 29

probability = norm.cdf(threshold, loc=mean, scale=std)
print(f"Вероятность того, что случайно выбранная порция эспрессо будет содержать менее 29 мл кофе:{probability}")

if probability <= 0.05:
    print(f"Реальная практика утверждениям кафе не соответствует")
else:
    print(f"Реальная практика утверждениям кафе соответствует")
```

Задача 2 - Вероятность распределения и уровень значимости

Фармацевтическая компания разработала новое лекарство, предназначенное для снижения уровня холестерина. По результатам предварительных исследований, утверждается, что среднее снижение уровня холестерина после применения лекарства составляет 20 мг/дл. Однако, на основании многолетних наблюдений за эффективностью аналогичных препаратов, известно, что снижение уровня холестерина подчиняется нормальному распределению с средним значением 18 мг/дл и стандартным отклонением 4 мг/дл. Врач хочет проверить, какова вероятность того, что применение нового лекарства приведет к снижению уровня холестерина на 22 мг/дл или больше, что будет считаться значительным улучшением по сравнению со стандартными лекарствами. Используя пороговое значение 0.01 определите такое снижение холестерина случайность или такого уровня снижения не может быть.

```python
mean = 18
std = 4
threshold = 22

probability = 1 - norm.cdf(threshold, loc=mean, scale=std)
print(f"Вероятность того, что применение нового лекарства приведет к снижению уровня холестерина на 22 мг/дл или больше:{probability}\n")

if probability <= 0.01:
    print(f"Такое снижение холестерина случайность")
else:
    print(f"Такого уровня снижения не может быть")
```

Задача 3 - Вероятность распределения

Рекламное агентство заявляет, что каждая печатная брошюра содержит ровно 20 страниц. Однако, из прошлых данных о качестве печати известно, что количество страниц в брошюре подчиняется нормальному распределению с средним значением 20 страниц и стандартным отклонением 2 страницы. Менеджер по контролю качества хочет проверить, какова вероятность того, что случайно выбранная брошюра будет содержать менее 18 страниц, что будет считаться недопустимым для стандартов агентства.

```python
probability = norm.cdf(18,loc=20, scale=2)
print(f"Вероятность того, что случайно выбранная брошюра будет содержать менее 18 страниц:{probability}")
```

Задача 4 - Центральная предельная теорема

Создайте свой список с случайными числами, пусть значения будут из нормального распределения, равномерного распределения и экспоненциального распределения, как на уроке.

● посчитайте среднее значение

● создайте рандомную выборку из 300 значений, посчитайте ее среднее значение

● постройте гистограмму и qq.plot

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

np.random.seed(101)

data = np.random.normal(50,5,300).tolist() + np.random.uniform(0,100,300).tolist() + np.random.exponential(10,300).tolist()

print(f"Среднее значение:{np.mean(data)}\n")

sample = np.random.choice(data, 300, False)

print(f"Среднее значение выборки:{np.mean(sample)}\n")

plt.hist(data,rwidth=0.9,density=True,alpha=0.4,label='Список')
plt.hist(sample,rwidth=0.9,density=True,alpha=0.4,label='Выборка')
plt.title('Гистограмма')
plt.xlabel('Значения')
plt.ylabel('Плотность вероятности')
plt.legend()

sm.qqplot(np.array(data),line='r')
plt.title('Список')
sm.qqplot(np.array(sample),line='r')
plt.title('Выборка')
plt.show()
```

Задание 5 - Центральная предельная теорема

Исследование показало, что среднее время, которое жители определённого города тратят на дорогу до работы, составляет 30 минут. Вы хотите проверить, насколько вероятно, что выборка из 50 случайных жителей этого города будет иметь среднее время в пути до работы, близкое к этому значению. (стандартное отклонение выборки = 10).



```python
from scipy.stats import norm
import numpy as np

# Заданные параметры
population_mean = 30 # среднее время в пути до работы в городе, минуты
std_dev = 10 # предполагаемое стандартное отклонение времени в пути, минуты
n = 50 # размер выборки

# Расчет стандартной ошибки среднего (SE)
SE = std_dev / np.sqrt(n)

# Определение вероятности, что среднее время в пути для выборки равно или меньше 30 минут
# Здесь используем CDF для нормального распределения с центром в population_mean и стандартной ошибкой SE
probability = norm.cdf(30, population_mean, SE)

print(f"Вероятность, что среднее время в пути до работы для выборки из 50 жителей равно или меньше 30 минут: {probability:.4f}")
```

Задача 6 - Оценка доверительного интервала

Вы исследуете средний рост студентов университета. Вы измерили рост случайно выбранных студентов и хотите оценить, насколько реальное распределение роста соответствует вашим наблюдениям. Вы предполагаете, что рост студентов нормально распределен с средним значением 175 см и стандартным отклонением 6 см. Оцените 95% доверительный интервал для среднего роста студентов и визуализируйте распределение с помощью KDE-графика.

```python
np.random.seed(101)
import seaborn as sns

data = np.random.normal(175,6,3000)
sns.kdeplot(data)
ul = norm.ppf(0.975, 175, 6)
ll = norm.ppf(0.025, 175, 6)
plt.axvline(ul, linestyle='--', color='red')
plt.axvline(ll, linestyle='--', color='red')
plt.title('95% доверительный интервал для среднего роста студентов ')
plt.xlabel('рост студентов')
plt.show()
```

Задача 7 - Точечная оценка

Нужно смоделировать распределение оценок студентов по определенному предмету. Из предыдущих лет известно, что средняя оценка по этому предмету составляет 75 баллов с стандартным отклонением 10 баллов. Мы хотим сгенерировать выборку из 100 оценок, чтобы проанализировать, как могут распределяться оценки текущего года. Посчитайте точечную оценку.



```python
np.random.seed(101)
data = np.random.normal(75,10,100)
norm_dist_std = np.std(data, ddof=1)/ np.sqrt(len(data))

print(f"Точечная оценка среднего значения:{data.mean()}")
print(f"Точечная оценка медианы:{np.median(data)}")
print(f"Точечная оценка отклонения:{norm_dist_std}")
```
