```python
#1. Создайте функцию f(x) = 3 * x^2 + 2 * x - 10
```

```python
def f(x):
    return 3 * x**2 + 2 * x - 10
```

```python
#2. Создайте список x_points (или массив через numpy), в котором у нас лежат числа от -10 до 10
```

```python
import numpy as np

x_points = np.arange(-10,11)
print(x_points)
```

```python
#3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения 
# x_points, также сохраните результаты в y_points (cписок или массив)
```

```python
y_points = []
for x in x_points:
    y = f(x)
    y_points.append(y)
    print(y)
```

```python
#4. Покажите на графике x_points & y_points
```

```python
import matplotlib.pyplot as plt

plt.plot(x_points,y_points, color='green', linestyle='--',marker='o',label='f(x)')
plt.legend()
plt.show()
```

```python
#5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)
```

```python
import sympy as sym

x = sym.Symbol('x')
f = sym.diff(3 * x**2 + 2 * x - 10)
print(f)

def f_diff(x):
    return 6*x + 2
```

```python
#6 Чему равна производная при x = 5
```

```python
print(f_diff(5))
```

```python
#7. Покажите на графике f(x) & f_diff(x)
```

```python
y1_points = []
for x in x_points:
    y = f_diff(x)
    y1_points.append(y)
plt.plot(x_points,y_points, color='green', linestyle='--',marker='o',label='f(x)')
plt.plot(x_points,y1_points, color='blue', linestyle=':',marker='o',label='f_diff(x)')
plt.legend()
plt.show()
```

```python
x = [i for i in range(0,6)]
y = []
for i in x:
    y.append(i**3 + i**2 - 10*i)
    
plt.plot(x,y, marker='o', linestyle=':',label='f_i(x)',color='green')
plt.axhline(color='black')
plt.fill_between(x,y,where = [(i>=0) and (i<=5) for i in x],color='blue',alpha=0.3)
plt.show()
```
