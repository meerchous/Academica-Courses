```python
#1. Создайте функцию f(x) = x^6 + 10 * x^3 - 1000
```

```python
def f(x):
    return x**6 + 10 * x**3 - 1000
```

```python
#2. Создайте список x_points (или массив через numpy), в котором у нас лежат числа от -10 до 10
```

```python
import numpy as np

x_points = np.arange(-10,10)
print(x_points)
```

```python
#3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения 
# x_points, также сохраните результаты в y_points (cписок или массив)
```

```python
y_points = []
for x in x_points:
    y = f(x)
    y_points.append(y)
    print(y)
```

```python
#4. Покажите на графике x_points & y_points
```

```python
import matplotlib.pyplot as plt

plt.plot(x_points,y_points, color='green', linestyle='--', label='linear function', marker='|')
plt.legend()
plt.show()
```

```python
#5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)
```

```python
import sympy as sym

x = sym.Symbol('x')
f = sym.diff(x**6 + 10 * x**3 - 1000)

def f_diff(x):
    return 6*x**5 + 30*x**2
    
```

```python
#6 Чему равна производная при x = 10
```

```python
print(f_diff(10))
```

```python
#7 Покажите на графике f(x) & f_diff(x)
```

```python
y1_points = []
for x in x_points:
    y = f_diff(x)
    y1_points.append(y)


plt.plot(x_points,y_points, color='green', linestyle='--', label='f(x) function', marker='o')
plt.plot(x1_points,y1_points, color='red', linestyle='-.', label='f_diff(x) function', marker='o')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

x_points =[]
y_points=[]
for i in range(-10,11):
    x_points.append(i)
    y_points.append(i**6 + 10 * (i**3) - 1000)
print(x_points)
print(y_points)



y_derivative=[]
for i in range(-10,11):
    y_derivative.append(6*(i**5)+30*(i**2))

print(y_derivative)

plt.plot(x_points, y_points, marker = 'o', linestyle = '--', color = 'blue', label='f(x)')
plt.plot(x_points,y_derivative, marker = 'o', linestyle = 'solid', color = 'green', label='f_diff(x)')
plt.legend()
plt.show()
```

```python
#8 Найдите уравнение касательной линии в точке x = 9
```

```python
x1 = 9
y1 = f(x1)
m = f_diff(x1)

print(f'y={m}*(x-{x1})+{y1}') 
```

```python
#9 Покажите на графике функцию f(x) & касательную линию в точке x = 9
```

```python
xk_points = [i for i in range(5,14)]
yk_points = []
for x in xk_points:
    y = m*(x-x1)+y1
    yk_points.append(y)


plt.plot(x_points,y_points, color='green', linestyle='--', label='f(x) function', marker='o')
plt.plot(xk_points,yk_points, color='red', linestyle='-.', label='fk(x) function', marker='o')
plt.legend()
plt.show()
```
