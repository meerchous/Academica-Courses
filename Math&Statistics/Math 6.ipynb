**Упражнение 1:** Рассмотрим систему уравнений, приведенную ниже. Найдите такое значение коэффициента $a$, чтобы система имела ровно одно решение (то есть найти любое подходящее значение $a$). Для этого значения $a$ найдите решение системы уравнений и начертите линии, образованные вашими уравнениями, чтобы проверить свой ответ.

$$
\begin{eqnarray*}
ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

a*x1+2*x2=1 --> (1)
4*x1+8*x2=12 --> (2)

(2)-4*(1):

(4-4*a)*x1=8
a = 0

x2 = (1-a*x1)/2 --> x2 = 1/2
x2 = (12-4*x1)/8

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
plt.plot(x,(1-0*x)/2)
plt.plot(x,(12-4*x)/8)

plt.text(-4,0.6,'$2x_2=1$')
plt.text(-2,2.7,'$4x_1+8x_2=12$')

plt.grid(True,ls=':')
```

**Упражнение 3:** Вычислите $-2E$, $G+F$, $4F-G$, $HG$ и $GE$, используя следующие определения матриц. Сначала выполните упражнение на бумаге, а затем проверьте, выполнив расчет с массивами NumPy. Для умножения матриц используйте np.dot

$$
\begin{equation}
E = \left[ \begin{array}{r} 5 \\ -2 \end{array}\right] \hspace{1cm} 
F = \left[ \begin{array}{rr} 1 & 6 \\ 2 & 0 \\ -1 & -1 \end{array}\right] \hspace{1cm}
G = \left[ \begin{array}{rr} 2 & 0\\ -1 & 3 \\ -1 & 6 \end{array}\right] \hspace{1cm}
H = \left[ \begin{array}{rrr} 3 & 0 & 1 \\ 1 & -2 & 2 \\ 3 & 4 & -1\end{array}\right]
\end{equation}
$$

```python
E = np.array([[5],[-2]])
F = np.array([[1,6],[2,0],[-1,-1]])
G = np.array([[2,0],[-1,3],[-1,6]])
H = np.array([[3,0,1],[1,-2,2],[3,4,-1]])


print(f'-2E: \n{-2*E}')
print(f'G+F= \n{G+F}')
print(f'4F-G= \n{4*F-G}')
print(f'HG= \n{np.dot(H,G)}')
print(f'GE= \n{np.dot(G,E)}')
```
