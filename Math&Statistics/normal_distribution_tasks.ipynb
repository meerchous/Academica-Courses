Задание 1 - Исследование нормальных распределений

* Сгенерируйте первый список из 1000 случайных чисел, следуя стандартному нормальному распределению (μ=0, σ=1).
* Сгенерируйте второй список из 1000 случайных чисел, следуя нормальному распределению с средним значением μ=8 и стандартным отклонением σ=2.
* Постройте гистограммы для обоих списков. Гистограммы должны быть нормализованы, чтобы отражать плотность вероятности.
* Оцентрируйте и измените масштаб нормального распределения с μ=8 и σ=2 так, чтобы его среднее значение стало 0, а стандартное отклонение 1.
* Постройте гистограмму для модифицированных данных, сравнивая её с гистограммой стандартного нормального распределения.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels as sm

list1 = np.random.normal(0,1,1000)
list2 = np.random.normal(8,2,1000)

plt.hist(list1,rwidth=0.9,density=True,alpha=0.5,label='стандарт нормальное распределение',bins = np.arange(-5,16))
plt.hist(list2,rwidth=0.9,density=True,alpha=0.5,label='нормальное распределение',bins = np.arange(-5,16))
plt.title('Первая гистограмма')
plt.xlabel('значения')
plt.ylabel('плотность вероятности')
plt.legend()
plt.show()
```

```python
list2 = list2 - list2.mean()
std= np.std(list2)
list2=list2/std

plt.hist(list1,rwidth=0.9,density=True,alpha=0.5,label='стандарт нормальное распределение',bins = np.arange(-5,16))
plt.hist(list2,rwidth=0.9,density=True,alpha=0.5,label='нормальное распределение',bins = np.arange(-5,16))
plt.title('Модифицированная гистограмма')
plt.xlabel('значения')
plt.ylabel('плотность вероятности')
plt.legend()
plt.show()
```

Задание 2 - Изучение распределения роста учащихся средней школы и его стандартизация для анализа.

* Используйте Numpy для создания выборки из 1000 случайных значений роста учащихся, следующих нормальному распределению с средним значением 170 см и стандартным отклонением 10 см.
* Преобразуйте исходные данные так, чтобы получить распределение с средним значением 0 и стандартным отклонением 1, используя простую стандартизацию.
* Постройте гистограммы для исходных данных роста учащихся и для стандартизированных данных.

```python
data = np.random.normal(170,10,1000)
print(data.mean())
print(data.std())


data1 = data - data.mean()
data1 = data1/data1.std()
print(f"средним значение: {data1.mean()}")
print(f"стандартным отклонением: {data1.std()}")

fig, axs = plt.subplots(1,2, figsize=(10,5))


axs[0].hist(data,rwidth=0.9,density=True,alpha=0.5,label='Исходные данные')
axs[0].set_title('Гистограмма №1')
axs[0].set_xlabel('рост')
axs[0].set_ylabel('плотность вероятности')
axs[0].legend()

axs[1].hist(data1,rwidth=0.9,density=True,alpha=0.5,label='Модифицированные данные')
axs[1].set_title('Гистограмма №2')
axs[1].set_xlabel('рост')
axs[1].set_ylabel('плотность вероятности')
axs[1].legend()
plt.show()
```

Задание 3 - Изменение климата приводит к тому, что погодные условия становятся более переменчивыми, что включает изменения в среднесуточных температурах. В этом задании вы будете работать с моделью данных среднесуточных температур для определенного месяца, например, июля. Исходя из исторических данных, предположим, что средняя температура в июле составляет 22°C со стандартным отклонением 3°C.

* Сгенерируйте выборку из 1000 случайных значений среднесуточных температур для июля, используя нормальное распределение с заданными параметрами (среднее = 22°C, стандартное отклонение = 3°C).
* Определите, какой процент дней характеризуется аномально высокими (выше 27°C) и аномально низкими (ниже 15°C) температурами.
* Постройте гистограмму среднесуточных температур для июля. Выделите области аномальных температур на гистограмме разными цветами.

```python
july = np.random.normal(22,3,1000)

hot = [val for val in july if val>27]
cold = [val for val in july if val<15]
print(f"процент дней с аномально высокими (выше 27°C):{len(hot)/len(july)*100}%")
print(f"процент дней с аномально низкими (ниже 15°C): {len(cold)/len(july)*100}%")

plt.hist(july,bins=np.arange(1,32,1),density=True,rwidth=0.9,color='lime')
plt.axvline(27,linestyle='--',color='red',label='аномальная жара')
plt.axvline(max(july),linestyle='--',color='red')
plt.axvline(15,linestyle='--',color='blue',label='аномальный холод')
plt.axvline(min(july),linestyle='--',color='blue')
plt.title('Среднесуточные температуры для июля')
plt.xlabel('Температура')
plt.ylabel('Плотность вероятности')
plt.legend(loc=2)
plt.show()
```
